### 1.1. Information about Dataset
The dataset contains analysis of wines grown in a certain region in Italy. The wines are made from 3 different varieties of grapes. There are 13 features showed on 3 varieties of wines. These are
1. Alcohol
2. Malic acid
3. Ash
4. Alcalinity of ash  
5. Magnesium
6. Total phenols
7. Flavanoids
8. Nonflavanoid phenols
9. Proanthocyanins
10. Color intensity
11. Hue
12. OD280/OD315 of diluted wines
13. Proline

Also the first column in dataset shows the types of the wines. 
With this information the updated dataframe will be as follows:

```python
cnames = ['class', 
          'alcohol', 
          'malic_acid', 
          'ash', 
          'ash_alcalinity', 
          'magnesium', 
          'ttl_phenols', 
          'flavanoids', 
          'nonflavanoid_phenols', 
          'proanthocyanins', 
          'color_intensity',
          'hue',
          'OD280/OD315',
          'prolin']
dfwine = pd.read_csv('Data/wine.csv', names=cnames)
dfwine.sample(10)
```

```python
print(dfwine.info())
dfwine.describe()
```

The information shows that there are no missing data and all of our data is numeric. Also, it looks like there is no error in our data apart from `ash` column. The minimum shows -1. After a small search on the web, I learnt that ash shouldn't be under zero. We have to clean that. 
## 2. Cleaning

```python
print(dfwine[dfwine.ash<0].ash.count())
```

There 30 entries below zero. If we exclude these rows we will lose nearly 15% of our data. Filling it with the mean of column would be appropriate.

```python
posval = dfwine[dfwine.ash>0].ash

upd_ix = dfwine[dfwine.ash<0].index
dfwine.loc[upd_ix, 'ash'] = posval.mean()
```

```python
dfwine.describe()
```

Now our data seems clean and ordered.

## Model Building

```python
X = dfwine.iloc[:,1:]
y = dfwine.iloc[:,0]
```

```python
from sklearn.model_selection import train_test_split

train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.metrics import accuracy_score

mnames = ['Decision Tree', 'K Neighbours', 'Support Vector Machine', 'Gradient Boosting', 'Random Forest']
models = [DecisionTreeClassifier(), 
          KNeighborsClassifier(), 
          SVC(), 
          GradientBoostingClassifier(), 
          RandomForestClassifier()]

for name, model in zip(mnames, models):
    model.fit(train_X, train_y)
    pred = model.predict(val_X)
    acc = accuracy_score(val_y, pred)
    
    print('The accuracy of {}: {:.2f}'.format(name, acc))
```

```python
from sklearn.model_selection import cross_val_score

for name, model in zip(mnames, models):
    scores = cross_val_score(model, X, y, cv=5)
    print('The accuracy of {}: {:.2f}'.format(name, scores.mean()))
```

```python
from sklearn.model_selection import LeaveOneOut
loo = LeaveOneOut()

for name, model in zip(mnames, models):
    scores = cross_val_score(model, X, y, cv=loo)
    print('The accuracy of {}: {:.2f}'.format(name, scores.mean()))
```

```python
param_grid = {'n_estimators': [5, 10, 100, 150, 200], 'min_samples_split': [5, 10, 20, 25, 50, 75, 100]}

from sklearn.model_selection import GridSearchCV

grid = GridSearchCV(RandomForestClassifier(), param_grid, cv=5)
```

```python
grid.fit(train_X, train_y)
```

```python
grid.score(val_X, val_y)
```

```python
print("Best parameters: {}".format(grid.best_params_))
print("Best cross-validation score: {:.2f}".format(grid.best_score_))
print("Best estimator:\n{}".format(grid.best_estimator_))
```
